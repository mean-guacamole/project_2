```python
avocado_df=avocado_data_df.drop(['Unnamed: 0','year','4046','4225','4770','Total Bags','Small Bags','Large Bags','XLarge Bags','type','region'], axis=1)
```

```python
avocado_df['Date']=pd.to_datetime(avocado_df['Date'])
```

```python
avocado_df=avocado_df.rename(columns={'Total Volume':'totalvolume','AveragePrice':'averageprice','Date':'dt'},inplace = False)
```

```python
avocado_clean=avocado_df.groupby(['dt']).agg({'averageprice':'mean', 'totalvolume':'sum'}).reset_index()
```

```python
avocado_clean
```

```python
trump_df=trump_data_df.drop(['Unnamed: 0'], axis=1)
```

```python
trump_clean=trump_df.rename(columns={'date':'dt'},inplace = False)
```

```python
trump_clean['dt']=pd.to_datetime(trump_clean['dt'])
```

```python
trump_clean=trump_clean.dropna()
```

```python
trump_clean
```

```python
avocado_clean.describe()
```

```python
trump_clean.describe()
```

```python
mean_guac=pd.merge(trump_clean,avocado_clean, on='dt')
```
